#Exercício 1

```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np

#img1 = cv.imread("blue_red.jpg")
img1 = cv.imread("teste.jpg")
img1 = cv.cvtColor(img1, cv.COLOR_BGR2RGB)

# CONVERTER a imagem de RGB para HSV
img2 = cv.cvtColor(img1, cv.COLOR_RGB2HSV)

#Matiz,Saturação,Luminância

# SEGMENTAR os pixels no intervalo da cor AZUL
ini = np.array([100,120,80])            
fim = np.array([170,255,255])
img3 = cv.inRange(img2,ini,fim)

# SEGMENTAR os pixels no intervalo da cor VERMELHO
ini = np.array([0,120,80])            
fim = np.array([20,255,255])
img4 = cv.inRange(img2,ini,fim)


# SEGMENTAR os pixels no intervalo da cor AMARELO
ini = np.array([20,120,80])            
fim = np.array([60,255,255])
img5 = cv.inRange(img2,ini,fim)

plt.figure(figsize=(12,12))
plt.subplot(141), plt.imshow(img1), plt.title('original'), plt.axis("off")
plt.subplot(142), plt.imshow(img3,cmap='binary'), plt.title('segmentação do azul'), plt.axis("off")
plt.subplot(143), plt.imshow(img4,cmap='binary'), plt.title('segmentação do vermelho'), plt.axis("off")
plt.subplot(144), plt.imshow(img5,cmap='binary'), plt.title('segmentação do amarelo'), plt.axis("off")
plt.show()
```

#Exercício 2

```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np

#img1 = cv.imread("blue_red.jpg")
img1 = cv.imread("lago.jpg")
img1 = cv.cvtColor(img1, cv.COLOR_BGR2RGB)

# CONVERTER a imagem de RGB para HSV
img2 = cv.cvtColor(img1, cv.COLOR_RGB2HSV)

#Matiz,Saturação,Luminância

ini = np.array([90,50,180])            
fim = np.array([170,255,255])
img3 = cv.inRange(img2,ini,fim)



plt.figure(figsize=(12,12))
plt.subplot(121), plt.imshow(img1), plt.title('original'), plt.axis("off")
plt.subplot(122), plt.imshow(img3,cmap='binary'), plt.title('area do lago'), plt.axis("off")
plt.show()
```
