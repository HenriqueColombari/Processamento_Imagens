```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
import random

!wget "https://homepages.cae.wisc.edu/~ece533/images/baboon.png"
```

```python
import random
img1 =cv.imread("baboon.png")
img1 = cv.cvtColor(img1, cv.COLOR_BGR2RGB) 

def sp_noise(image,prob):
    output = np.zeros(image.shape,np.uint8)
    thres = 1 - prob 
    for i in range(image.shape[0]):
        for j in range(image.shape[1]):
            rdn = random.random()
            if rdn < prob:
                output[i][j] = 0
            elif rdn > thres:
                output[i][j] = 255
            else:
                output[i][j] = image[i][j]
    return output

img2 = sp_noise(img1, 0.05)

img3 = cv.GaussianBlur(img2, (15,15), 0)

img4 = cv.medianBlur(img1, 9)

plt.figure(figsize=(15,15))
plt.subplot(131), plt.imshow(img2), plt.axis('off'), plt.title('Salt and Pepper')
plt.subplot(132), plt.imshow(img3), plt.axis('off'), plt.title('Gaussian Blur')
plt.subplot(133), plt.imshow(img4), plt.axis('off'), plt.title('Median Blur')
plt.show()
```
