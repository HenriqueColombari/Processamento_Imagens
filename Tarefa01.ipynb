```python
import cv2 as cv
import matplotlib.pyplot as plt

!wget "https://homepages.cae.wisc.edu/~ece533/images/peppers.png"
!wget "https://homepages.cae.wisc.edu/~ece533/images/baboon.png"
!wget "https://homepages.cae.wisc.edu/~ece533/images/lena.png"
```

```python
img1 =cv.imread("peppers.png")
img2 =cv.imread("baboon.png")
img3 =cv.imread("lena.png")

img1 = cv.cvtColor(img1, cv.COLOR_BGR2RGB)
img2 = cv.cvtColor(img2, cv.COLOR_BGR2GRAY)
img3 = cv.cvtColor(img3, cv.COLOR_BGR2GRAY)


for lin in range(img3.shape[0]):   
  for col in range(img3.shape[1]):
    
    pixel = img3.item(lin,col)
    if ( pixel > 127 ):
      img3.itemset( (lin,col), 255)
    else:
      img3.itemset( (lin,col), 0)


print("Quantidade de pixels =", img1.shape[0], "*", img1.shape[1], "=",  img1.shape[0]*img1.shape[1])

print("Quantidade de linhas da imagem 2 =", img2.shape[0])

print("Quantidade de colunas da imagem 3 =", img2.shape[1])

plt.figure(figsize=(10,10))
plt.subplot(131), plt.imshow(img1), plt.axis('off'), plt.title('original')
plt.subplot(132), plt.imshow(img2,cmap='gray'), plt.axis('off'), plt.title('gray')
plt.subplot(133), plt.imshow(img3,cmap='gray'), plt.axis('off'), plt.title('binária')
plt.show()
```

```python
import cv2 as cv
import matplotlib.pyplot as plt

!wget "https://homepages.cae.wisc.edu/~ece533/images/peppers.png"
!wget "https://homepages.cae.wisc.edu/~ece533/images/baboon.png"
!wget "https://homepages.cae.wisc.edu/~ece533/images/airplane.png"
!wget "https://homepages.cae.wisc.edu/~ece533/images/lena.png"
```

```python
img1 =cv.imread("peppers.png")
img2 =cv.imread("baboon.png")
img3 =cv.imread("airplane.png")
img4 =cv.imread("lena.png")

img1 = cv.cvtColor(img1, cv.COLOR_BGR2RGB)
img2 = cv.cvtColor(img2, cv.COLOR_BGR2RGB)
img3 = cv.cvtColor(img3, cv.COLOR_BGR2RGB)
img4 = cv.cvtColor(img4, cv.COLOR_BGR2RGB)

print("PLOT DAS IMAGENS ORIGINAIS")
plt.figure(figsize=(5,5))
plt.subplot(221), plt.imshow(img1), plt.axis('off')
plt.subplot(222), plt.imshow(img2), plt.axis('off')
plt.subplot(223), plt.imshow(img3), plt.axis('off')
plt.subplot(224), plt.imshow(img4), plt.axis('off')
plt.show()

img5 = img1.copy()
for i in range(img1.shape[0]):
  for k in range(img1.shape[1]):
    img5.itemset((i, k, 0), img1.item(i, k, 0) + img2.item(i, k, 0) + img3.item(i, k, 0)*0.5 + img4.item(i, k, 0) - img1.item(i, k, 0))
    img5.itemset((i, k, 1), img1.item(i, k, 1) + img2.item(i, k, 1) + img3.item(i, k, 1)*0.5 + img4.item(i, k, 1) - img1.item(i, k, 1))
    img5.itemset((i, k, 2), img1.item(i, k, 2) + img2.item(i, k, 2) + img3.item(i, k, 2)*0.5 + img4.item(i, k, 2) - img1.item(i, k, 2))



print("\nPLOT DO RESULTADO DAS OPERAÇÕES:")
plt.figure(figsize=(10,10))
plt.subplot(221), plt.imshow(img5), plt.axis('off')
plt.show()

img6 = img1.copy()
for i in range(img1.shape[0]):
  for k in range(img1.shape[1]):
    img6.itemset((i, k, 0), img1.item(i, k, 0) * -1)
    img6.itemset((i, k, 1), img1.item(i, k, 1) * -1)
    img6.itemset((i, k, 2), img1.item(i, k, 2) * -1)


print("\nPLOT DO RESULTADO DA INVERSÃO DA IMAGEM 1:")

hr1 = cv.calcHist([img1],[0],None,[256],[0,255])
hg1 = cv.calcHist([img1],[1],None,[256],[0,255])
hb1 = cv.calcHist([img1],[2],None,[256],[0,255])

hr6 = cv.calcHist([img6],[0],None,[256],[0,255])
hg6 = cv.calcHist([img6],[1],None,[256],[0,255])
hb6 = cv.calcHist([img6],[2],None,[256],[0,255])

plt.figure(figsize=(5,5))
plt.subplot(221), plt.imshow(img1), plt.axis('off')
plt.subplot(222), 
plt.plot(hr1,'-',color='red'), plt.xlim(0,255),plt.ylim(0,max(hr1)+100), plt.axis('off'),
plt.plot(hg1,'-',color='green'), plt.xlim(0,255),plt.ylim(0,max(hg1)+100),
plt.plot(hb1,'-',color='blue'), plt.xlim(0,255),plt.ylim(0,max(hb1)+100), 

plt.subplot(223), plt.imshow(img6), plt.axis('off')
plt.subplot(224),
plt.plot(hr6,'-',color='red'), plt.xlim(0,255),plt.ylim(0,max(hr6)+100), plt.axis('off'),
plt.plot(hg6,'-',color='green'), plt.xlim(0,255),plt.ylim(0,max(hg6)+100),
plt.plot(hb6,'-',color='blue'), plt.xlim(0,255),plt.ylim(0,max(hb6)+100), 

plt.show()

```

```python
import cv2 as cv
import matplotlib.pyplot as plt

!wget "https://homepages.cae.wisc.edu/~ece533/images/peppers.png"
!wget "https://homepages.cae.wisc.edu/~ece533/images/baboon.png"
```

```python
img1 =cv.imread("peppers.png")
img2 =cv.imread("baboon.png")

img1 = cv.cvtColor(img1, cv.COLOR_BGR2RGB)
img2 = cv.cvtColor(img2, cv.COLOR_BGR2RGB)

img3 = img1.copy()
for i in range(img1.shape[0]):
  for k in range(img1.shape[1]):
    img3.itemset((i, k, 0), img1.item(i, k, 0)*0.8  + img2.item(i, k, 0)*0.2)
    img3.itemset((i, k, 1), img1.item(i, k, 1)*0.8  + img2.item(i, k, 1)*0.2)
    img3.itemset((i, k, 2), img1.item(i, k, 2)*0.8  + img2.item(i, k, 2)*0.2)


plt.imshow(img3), plt.axis('off')
plt.show()
```
